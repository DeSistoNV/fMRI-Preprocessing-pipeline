### Step 3: Specify parameters for the preprcessing pipeline

Now that the database has been created, we are ready to start pre-processing some data. As mentioned, we do data pre-processing using nipype. nipype is a python pacakge that makes it (sort of) easy to create and execute fairly complicated pipelines. Before proceeding, go see nipype and work through the [documentation](http://nipy.sourceforge.net/nipype/documentation.html). Come back once you have some idea what nipype is.

...

Okay, ready? In our processing pipeline, the root node--or *input node*--aggregates all of the information needed to run the pipeline on a bunch of data. The node is implemented as a subclass of the [nipype.IdentityInterface](http://nipy.sourceforge.net/nipype/interfaces/generated/nipype.interfaces.utility.html) node.

During execution of the pipeline, other nodes responsible for specific pre-processing stages are fed their inputs and parameter values from the input node. In summary, the pipeline is like this:






```python
Image(filename='workflow.png')
```

Therefore, to run the nipype pipeline, we must first construct an input node (the one labeled "query db" above) and supply it with values that it can farm out to the rest of the pipline.

```python
##create the input node 
main_inputnode = fsl_preproc_inode()

##note that it currently has no values assigned to any of the inputs that the pipeline will need
main_inputnode.inputs
```

Once the input node is created, we will first collect all of the datafiles that contain the imaging data we want to preprocess. 
We do this by specifying:
* **subject**: the subject
* **experiment**: the experiment id
* **run_list**: a list of runs we want to process. these are just integers indicating runs for a specific subj/experiment. If we want all of the runs, we just pass an empty list (this is almost always want we want, unless we are just trying to debug and to grab a subset).
* **sess_list**: a list of sessions. again, an empty list grabs them all. a list like [0,1] would just get the first two sessions.
* **crop_this**: a rule for cropping volumes. this is a nasty but necessary parameter, because sometimes people make mistakes. basically, everytime an experiment is run, the user has to tell the scanner the total number of voxels to collect. when the scanner operator isn't perfectly consistent, we can end up with different numbers of voxels across sessions/runs. we don't want to throw data away, so we have to crop the larger of the volumes to match the smaller. We do this specify a rule for cropping each dimension. The options for cropping are:
    - *min*: crop from left to right
    - *max*: crop from right to left
    - *med*: crop from center-out
    - So, we create a list like ['min', 'min', 'max'] the indicates how to crop along each dimension.

```python
##example: collect all the data we have for subject TN, imagery receptive field experiment
pipeline_data_params = dict()
pipeline_data_params['subject'] = 'TN'
pipeline_data_params['experiment'] = 'imagery.rf'
pipeline_data_params['run_list'] = []
pipeline_data_params['sess_list'] = []
pipeline_data_params['crop_this'] = ['med', 'med', 'min']  ##note that this will almost never be needed, and isn't need in this example

```

```python
##tell the inputnode the database it needs to query and open it
pipeline_data_params['db'] = my_db
main_inputnode.open_db(pipeline_data_params['db'])

```

```python
##now pass the fsl_preproc_params dictionary to the input node so it can collect all the data it needs
main_inputnode.populate_fields(subj=pipeline_data_params['subject'],
			       expID=pipeline_data_params['experiment'],
			       runList = pipeline_data_params['run_list'],
			       sessList = pipeline_data_params['sess_list'],
			       cropRule = pipeline_data_params['crop_this'])	  	##we want all runs here.
main_inputnode.close_db()
```

```python
##now some of the inputs have been acquired. the "populate fields" method does a lot of work
main_inputnode.inputs
```

You'll note above that many of the values of the input variables remain undefined. These are variables that are *not* stored in the database. Rather, they are input parameters for the many [alignment functions](http://fsl.fmrib.ox.ac.uk/fslcourse/lectures/reg.pdf) (mostly from FSL) that nipype will call while running the pipeline. Here they are:

**GENERAL**
* **basedir**: the pipeline creates *tons* of intermediate outputs that you don't need to worry about. these go here.
* **results_base_dir**: base directory (e.g., the hardrive) where the output is dumped
* **results_container**: the specific directory in the base_dir where the data dumps
* **convert_dicoms**: if we've already converted to nifti, put False. otherwise, go for it, but note problems mentioned above.
* **t_size**: a terrible hack. this is the largest number of volumes you could ever expect to obtain in one run. so just set it to 10^6 and don't ask any questions.
* **nProc**: nipype will run on a cluster. if your machine has multiple cores, enter that number here for maximum speed. if you are debugging, put 1 here. or go insane. whatever you prefer.

**SKULL STRIPPING**

* **bet_frac**: bet = FSL's brain extraction tool. It creates a mask for the brain (does skull stripping, etc). here is a sensitivity threshold. the mask will come out at the end of the pipeline, so you will be able to tell if you mess this one up.

**MOTION CORRECTION**

* **ref_run**: the reference volume that all volumes will be aligned to. if left empty, use first vol of first run.
* **moco_only**: apply motion correct only. if True, all other parameters regarding alignment (e.g., FLIRT and FNIRT params) are ignored.

**LINEAR REGISTRATION**

* **FLIRT_cost_fun**: you have options, but the only joy I've ever gotten is using 'normcorr'.
* **interp_FLIRT**: as below for FNIRT, but for the linear registration stage [FLIRT](http://fsl.fmrib.ox.ac.uk/fsl/fslwiki/FLIRT/UserGuide)
* **dof_FLIRT**: degrees of freedom for linear stage. max is 12 (translation (3) , rotation (3), scaling (1), shearing (4?)
* **rigid2D_FLIRT**: If true, restrict linear registration to rigid-body transformations. If true, ignore the dof_FLIRT param.

**NONLINEAR REGISTRATION**

* **do_FNIRT**: if false, will run FLIRT only (unless moco_only = True)
* **interp_FNIRT**: what flavor of interpolation would you like to use after nonlinear warping? your choices are "spline" or "sinc". "spline" is better for debugging, "sinc" is better for the final version of your data.
* **searchr_x, y, and z**: parameters for nonlinear registration [FNIRT](http://fsl.fmrib.ox.ac.uk/fsl/fslwiki/FNIRT). Here they, are exposed for you use. Don't remember what they do. If you find out, tell me. Otherwise, leave empty and use the FSL defaults (whatever they are).
* **FNIRT_subsamp**: FNIRT runs a coarse-to-fine algorithm. This is a list specifying the downsampling factor on each iteration.
* **FNIRT_warpres**: Resolution of the warping function. Like, how fine is the warping. Can specify different level for each iteration. *Question*: Why is this list shorter than the one above? Check [this](http://fsl.fmrib.ox.ac.uk/fslcourse/lectures/reg.pdf) to find out.

I sometimes find it helpful to create a dictionary for all these parameters, because I sometimes want to pickle it for future use:

```python
fsl_preproc_params = dict()
##general
fsl_preproc_params['basedir'] = '/Data/tmp/Imagery_RF_test/'
fsl_preproc_params['results_base_dir'] = '/Data/tmp/'
fsl_preproc_params['results_container'] = 'Imagery_RF_test'
fsl_preproc_params['convert_dicoms'] = False
#default_params['ref_vol_runList'] = [0]							##grab only the first run for aligning all the other runs to
fsl_preproc_params['t_size'] = 10000
fsl_preproc_params['nProc'] = 1
fsl_preproc_params['bet_frac'] = 0.1 ##<<seems to work


##motion correction
fsl_preproc_params['ref_run'] = [] 
fsl_preproc_params['moco_only'] = False

##linear registration
fsl_preproc_params['interp_FLIRT'] = 'sinc'
fsl_preproc_params['dof_FLIRT'] = 12
fsl_preproc_params['rigid2D_FLIRT'] = False

##nonlinear registration
fsl_preproc_params['searchr_x'] = []
fsl_preproc_params['searchr_y'] = []
fsl_preproc_params['searchr_z'] = []
fsl_preproc_params['interp_FNIRT'] = 'spline'
fsl_preproc_params['do_FNIRT'] = True
fsl_preproc_params['FNIRT_subsamp'] = [[4,2,1,1]]
fsl_preproc_params['FNIRT_warpres'] = [(5,5,5)]
fsl_preproc_params['FLIRT_cost_func'] = ['normcorr']#
```

### Step 4: Run the pipeline

The pipeline itself is constructed when we pass the input node to the function called *create_fsl_preproc_workflow.py*. Go and look at it. You will see it has two main sections. In one section, we create nodes corresponding to each of the pre-processing stages (the circles in the diagram above; note that there more nodes than pictured above, a complete diagram is below). These nodes are nipype wrappers--called [*interfaces*](http://nipy.sourceforge.net/nipype/users/interface_tutorial.html)--to the FSL functions we'll be using. The second section of the function connects the nodes to each other, grabbing input parameters from the input node as necessary. To create the workflow, we simply write:



```python
##====construct the workflow
pp = create_fsl_preproc_workflow(main_inputnode)

##and, for reasons I can't remember, establish the base directory as an attribute of the workflow
pp.base_dir = fsl_preproc_params['basedir']
```

Finally, we run the damn thing. In general, you'll want to check to see if multiple processors are requested, or just one.

If we've done everything correctly, the code below should execute cleanly. For typical datasets it will take < 10 hrs. Sometimes less than one hour. 

Chances are first few (or many) times you try to run this, it will crash. See the box just after the next for debugging tips.

```python
if fsl_preproc_params['nProc'] > 1:
  pp.run(plugin='MultiProc', plugin_args={'n_procs' : fsl_preproc_params['nProc']})
else:
  pp.run()
```

### Step 4.5: Debug.

If you're reading this, the code has crashed. Big surprise.

When a nipype pipeline crashes, nipype really spills it guts. The error messages are usually not super helpful. However, they will usually refer to specific nodes, and try to tell you what problems the nodes might have had when they we called. Check to full workflow diagram below, as well as the *create_fsl_preproc_workflow.py* source code to figure out what these nodes do and what might be wrong with them.

More importantly, whenever a pipeline crashes nipype creates a "crashdump" file. The very final error message will tell you where this file is located. Find that file and do the following:

    crashinfo = loadpkl(your_crashdump_file)
    %pdb                     #starts the debugger
    crashinfo['node'].run()  #re-runs the process. when it crashes, the debugger takes over
    pdb> up                  #typically, but not necessarily, the crash is one stack frame up
    pdb> your work here      #<<<inspect variables and fix stuff
    pdb> quit                 #close the debugger


### Step 5: Check the results

You know where the results are because you told nipype to put them into *results_base_dir+results_container*.

The last node in our pipeline is a [datasink](http://nipy.sourceforge.net/nipype/users/grabbing_and_sinking.html). We've programmed this node to distribute outputs across separate folders:

* **brain_mask**: this a binary volume that masks all things brain.
* **corrected_movie**: this contains a "movie" of the time-averaged volume for each run *after correction*. So if we had five runs, you'll find in this folder .nii file containing five volumes. read it into FSL view and check to see how well the pipeline worked.
* **uncorrected_movie**: as above, but before application of the pipeline. compared to the corrected_movie to get a sense for how much better the data look.
* **final_aligned**: this is the data you will use for analysis. each run is in it's own folder called *"_working_volxx"*.

### Appendix: detailed workflow diagram and description.

See the workflow diagram below. This is for the most complicated case where you have multiple runs and aligning them across multiple sessions requires some non-linear registration. 

Each ellipsoid is a node. We've talked about the input node above.

Each box is a subflow.

Description of subflows:
* **conv2nii**: converts dicom to nifti
* **killref**: if there is a reference image in the series (this is indicated in the database), remove it before processing
* **padvol**: if there are too many or too few volumes in this series (because perhaps you miscalculated the number of volumes you’d need before starting the run), this will either remove the extras, or add a few missing volumes. Missing volumes are replaced by the mean of the whole run. Obviously, be careful with this.
* **meanmerge**: this concatenates the mean image of all the runs in a series. there is an “uncorrected” and a “corrected” version that are used to check the quality of cross-run registration.

The rest of the nodes have to do with applying and managing alignment. Why so complicated? My goal was to be able to combine multiple rounds of alignment (if necessary). For example, when we have many runs that we acquired on separate days, we may need to apply some nonlinear registration on top of the motion correction. Also, nonlinear registration requires linear registration as a first step. We don’t want to interpolate every time we transform. So, we don’t want to have to do motion correct →  interpolate → linear registration → interpolate → nonlinear register → interpolate. Instead we do the following:

1. brain extraction (bet_vols)
2. send the extracted volumes to motion correction (moco_vols) and retain the xform and motion-corrected 4D volumes.
3. meanwhile, send the motion corrected volumes to a function that splits the 4D volumes up into a bunch of 3D volumes (split.Split.fsl). This is necessary because for FLIRT and FNIRT you can only apply transformations--or *xforms*--to 3D volumes.
4. Send the list of motion corrected 3D volumes to FLIRT. Apply FLIRT and retain the a linear transform parameters for each volume.
5. Send the list of 3D volumes to FNIRT along with the linear xform from FLIRT. Run FNIRT to get a nonlinear xform. 
6. Send the xform from FNIRT to applyWarp along with the original moco_vols xform. These are concatenated by applyWarp and then applied to each of the 3D volumes individually. Note there is an intermediate step called “expand_fnirt” (or if we are only doing FLIRT but not FNIRT it is called “expand_flirt”) that arranges the list of 3D volumes so that the moco xforms for separate runs are matched to the appropriate FNIRT (or FLIRT) xforms.
7. The split-up, aligned 3D volumes are now put back together into their appropriate runs, and remerged (“group_aligned_vol_list” and “remerge” nodes).
8. The means of each aligned run are merged to make a “corrected movie” that can be used to visually inspect how well the runs have been aligned across sesssions (compared to “uncorrected movie”).
9. The aligned volumes are re-named as “working_vols” by the data sink.
